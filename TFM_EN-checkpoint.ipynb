#### Venn Diagram

<div>In many cases, patients could receive combine treatments, for that reason a Venn Diagram will help to have a better approach than the individual plots.</div>
The first step, is to created the possible combinations, for example the variable `chemo` will be the result of  `chemotherapy` = <span style="color:green"> 1 </span>, `radio_therapy`= <span style="color:red">0</span> y `hormone_therapy` = <span style="color:red">0</span>.

<div>This step applies for each therapy, and then the possible combinations between them.</div>


````python
chemo = metabric[(metabric["chemotherapy"]==1) & (metabric["radio_therapy"]==0) & (metabric["hormone_therapy"]==0)]
radio = metabric[(metabric["chemotherapy"]==0) & (metabric["radio_therapy"]==1) & (metabric["hormone_therapy"]==0)]
hormonal = metabric[(metabric["chemotherapy"]==0) & (metabric["radio_therapy"]==0) & (metabric["hormone_therapy"]==1)]
chemo_radio = metabric[((metabric["chemotherapy"]==1) & (metabric["radio_therapy"]==1)) & (metabric["hormone_therapy"]==0)]
radio_hormonal = metabric[(metabric["chemotherapy"]==0) & ((metabric["radio_therapy"]==1) & (metabric["hormone_therapy"]==1))]
hormonal_chemo = metabric[((metabric["chemotherapy"]==1) & (metabric["hormone_therapy"]==1)) & (metabric["radio_therapy"]==0) ]
all_3 = metabric[((metabric["chemotherapy"]==1) & (metabric["radio_therapy"]==1) & (metabric["hormone_therapy"]==1))]

'''
Example on how Venn Diagram works:

value=[1,2,3,4,5,6,7]
labels=A,B,C

So:

A=1,B=2,A ∩ B=3, C=4, A∩C =5, B∩C =6, A∩B∩C = 7
'''

tratamientos = [chemo, radio, chemo_radio,hormonal, hormonal_chemo, radio_hormonal,all_3]
````

### Correlations between categorical features

At this point, using Pearson's method for categorical features is not the best approach, instead of that Cramer's V seems better.

````python
def cramers_V(var1,var2): 
    crosstab =np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) 
    stat = chi2_contingency(crosstab)[0] 
    obs = np.sum(crosstab)
    mini = min(crosstab.shape)-1 
    return (stat/(obs*mini))
````

To evaluate the metrics, a function call `mostrar_resultados` is called. It returns a confussion matrix, recall, precision and f1 and ROC Curve

* recall = indicate how good the model could detect a class.
* precision = indicate the quality of the model in classification task.
* f1 = combine the previous two

#### Metrics

The metrics to consider are confussion matrix, and f1 score.

Parameters: 

* Precision: measure the quality of the model in classification tasks. Their formula expresses the number of true positives divided number of positives.
* Recall: indicate how well the model is capable of detecting a class. It is expressed by the number of true positives divided number of cases that should be positive.
* F1 score: is a mix between precision and recall. It is used to measure the accuracy of the testing dataset.
* ROC Curve: indicate how good is the model classifying classes. The higher is the area under the curve, the better will be the results to identify the classes.
  <div>  On the y axis is the True Positive Rate, which is equal to the sensibility of the model and matches with the recall metric. On the x-axis is the False Positive Rate which corresponds to the specificity, the proportion of negatives cases diagnosticated as positives. </div>
   <div> The curve will take values between 0 and 1, values close to 1 have a better performance. </div>

````python
acc_train = []
acc_test = []
model = []
````

````python
def metricas(y_test, y_pred, y_train, models, X_train, X_test):
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 5))
    sns.heatmap(conf_matrix, annot=True, fmt="d");
    plt.title("Matriz confusión")
    plt.ylabel('Clase real')
    plt.xlabel('Clase predicha')
    plt.show()
    
    metrics.plot_roc_curve(models, X_test, y_test)
    model.append(models)
    acc_train.append(accuracy_score(y_train, models.predict(X_train)))
    acc_test.append(accuracy_score(y_test, y_pred))
    
    print (classification_report(y_test, y_pred))
    print('El accuracy para el conjunto de entrenamiento es de :', accuracy_score(y_train, models.predict(X_train)))
    print('El accuracy para el conjunto de testing es de :', accuracy_score(y_test, y_pred))
````

````python
def metricas_redes(y_test, y_pred, y_train, models, X_train):
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 5))
    sns.heatmap(conf_matrix, annot=True, fmt="d");
    plt.title("Matriz confusión")
    plt.ylabel('Clase real')
    plt.xlabel('Clase predicha')
    plt.show()
    
    model.append(models)
    acc_train.append(accuracy_score(y_train, models.predict_classes(X_train)))
    acc_test.append(accuracy_score(y_test, y_pred))
    
    print (classification_report(y_test, y_pred))
    print('El accuracy para el conjunto de entrenamiento es de :', accuracy_score(y_train, models.predict_classes(X_train)))
    print('El accuracy para el conjunto de testing es de :', accuracy_score(y_test, y_pred))
    
````

Confussion matrix consists in four values:

* TN = True Negative
* FP = False Positive
* FN = False Negative
* TN = True Negative
